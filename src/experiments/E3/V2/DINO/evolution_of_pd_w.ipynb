```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pd_w_results.csv")
df
```

```python
df.columns
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["frame_num"], df["pd_r"], color="red", label=r"The detection probability of the target $C_R$")
plt.plot(df["frame_num"], df["pd_l"], color="blue", label=r"The detection probability of the target $C_L$")
plt.plot(df["frame_num"], df["w_r"], color="orange", label=r"The weight of the target $C_R$")
plt.plot(df["frame_num"], df["w_l"], color="cyan", label=r"The weight of the target $C_L$")
plt.xlabel("Frame number")
plt.ylabel("Value")
plt.title("Evolution of targets' detection probabilities and weights")
plt.legend()
plt.savefig("pd_w_evolution.png")
```
