```python
import cv2
import os

import sys 



def get_image_size_in_bytes_cv2(img):
    # Read the image

    # Encode the image to PNG format
    _, img_encoded = cv2.imencode('.png', img)

    # Get the size of the encoded image in bytes
    size_bytes = len(img_encoded)
    
    return size_bytes


def resize_image(input_path, output_path, target_size_kb):
    # Read the image
    img = cv2.imread(input_path)

    # Calculate target size in bytes
    target_size_bytes = target_size_kb * 1024

    # Get image dimensions
    height, width, _ = img.shape

  
    #     img = cv2.resize(img, (int(width * 0.9), int(height * 0.9)))
#     img = cv2.resize(img, (0, 0), fx = 0.9, fy = 0.9)
    a = 0.9
    resized_img = img.copy()
    while True:
        resized_img=cv2.resize(img, (int(width*a), int(height*a)))
        resized_size = get_image_size_in_bytes_cv2(resized_img)/1000
        
        if resized_size > 300 and resized_size < 350:
            break
        
        if resized_size > target_size_kb:
            a-=0.01
        else:
            a+=0.01
        print(resized_size)
        
        # Write the resized image to output path
    cv2.imwrite(output_path, resized_img)

    print(f"Image resized and saved at {output_path} ")

def resize_images_in_folder(numerical_png_files, target_size_kb):
    for file_path in numerical_png_files:
        input_path = file_path
        output_path = f"{file_path[:-4]}.png"  # Convert to jpg format
        print(input_path, output_path)
        resize_image(input_path, output_path, target_size_kb)
    
    
    # Iterate over all files in the folder
#     for filename in os.listdir(folder_path):
#         Check if file is a PNG image with a numerical name
#         if filename.endswith(".png") and filename[:-4].isdigit():
#             input_path = os.path.join(folder_path, filename)
#             output_path = os.path.join(folder_path, f"{filename[:-4]}.png")  # Convert to jpg format
#             resize_image(input_path, output_path, target_size_kb)


def find_numerical_png_files(root_folder):
    numerical_png_files = []

    # Walk through all subfolders and files in the root folder
    for folder_path, _, files in os.walk(root_folder):
        for filename in files:
            # Check if the file is a PNG image with a numerical name
            if filename.endswith(".png") and filename[:-4].isdigit():
                # Append the file path to the list
                numerical_png_files.append(os.path.join(folder_path, filename))

    return numerical_png_files

# Example usage
root_folder = "./"  # Replace with the path to your root folder
numerical_png_files = find_numerical_png_files(root_folder)

target_file_size_kb = 400

resize_images_in_folder(numerical_png_files, target_file_size_kb)
```
