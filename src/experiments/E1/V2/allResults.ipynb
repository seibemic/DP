```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_sam = pd.read_csv("./SAM/sam_results.csv", index_col=0)
df_yolo = pd.read_csv("./YOLO/yolo_results.csv", index_col=0)
df_dino = pd.read_csv("./DINO/dino_results.csv", index_col=0)
df_spd = pd.read_csv("./noPd/noPd_results.csv", index_col=0)
```

```python
plt.figure(figsize=(12,4))
#plt.plot(df_sam["frame_num"], df_sam["sam_targets_in_queue"], color="red", label="SAM: all targets")
plt.plot(df_sam["frame_num"], df_sam["sam_targets_displayed"],color="red", label="SAM: displayed targets")

#plt.plot(df_yolo["frame_num"], df_yolo["yolo_targets_in_queue"], color="blue", label="SAM: all targets")
plt.plot(df_yolo["frame_num"], df_yolo["yolo_targets_displayed"],color="blue", label="YOLO: displayed targets")

#plt.plot(df_dino["frame_num"], df_dino["dino_targets_in_queue"], color="green", label="SAM: all targets")
plt.plot(df_dino["frame_num"], df_dino["dino_targets_displayed"],color="green", label="DINO: displayed targets")

plt.plot(df_spd["frame_num"], df_spd["staticPd_targets_displayed"],color="orange", label="Static Pd: displayed targets")

plt.plot(df_dino["frame_num"], df_dino["true_targets"],"-.",color="black", label="True targets count")
plt.xlabel("Frame number")
plt.ylabel("Number of targets")
#plt.yticks([0,1,2])
plt.title("Number of targets in time")
plt.legend()
plt.savefig("all_det.png",bbox_inches='tight')
plt.show()
```
