```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("dino_results.csv", index_col=0)
df
```

```python
plt.figure(figsize=(12,4))
plt.plot(df["frame_num"], df["dino_detects"],color="red", label="DINO detections")
plt.plot(df["frame_num"], df["dino_targets_in_queue"], color="orange", label="All targets in set")
plt.plot(df["frame_num"], df["dino_targets_displayed"],color="blue", label="Displayed targets")
plt.plot(df["frame_num"], df["true_targets"],"-.",color="black", label="True targets count")
plt.xlabel("Frame number")
plt.ylabel("Number of targets")
#plt.yticks([0,1,2])
plt.title("Number of targets in time")
plt.legend()
plt.savefig("dino_det.png",bbox_inches='tight')
plt.show()
```
