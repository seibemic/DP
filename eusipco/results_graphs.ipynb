```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("results.csv")
df
```

```python
plt.figure(figsize=(8,2))
plt.plot(df["time"], df["constant pd=0.95, 1 SP"],color="red", label="Pd=0.95, 1 SP")
plt.plot(df["time"], df["constant pd=0.7, 1 SP"], color="orange", label="Pd=0.7, 1 SP")
plt.plot(df["time"], df["dynamic pd"],color="blue", label="dynamic Pd")
plt.plot(df["time"], df["true count"],"-.",color="black", label="true count")
plt.xlabel("Time")
plt.ylabel("Number of targets")
plt.yticks([0,1,2])
plt.title("Number of targets in time")
plt.legend()
plt.savefig("1_sp.png",bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.plot(df["time"], df["dynamic pd unique id (max id)"],color="blue", label="dynamic Pd,  max id")
plt.plot(df["time"], df["constant pd=0.95, 2 SP, unique id (max id)"], color="red", label="Pd=0.95, 2 SP, max id")
plt.plot(df["time"], df["true count unique id (max id)"],"-.",color="black", label="true count, max id")
plt.xlabel("Time")
plt.ylabel("ID")
plt.yticks(range(0,7))
plt.title("Targets' max ID shown in time")
plt.legend()
plt.savefig("2_sp_id.png",bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,2))
plt.plot(df["time"], df["dynamic pd"], label="dynamic Pd")
plt.plot(df["time"], df["constant pd=0.95, 2 SP"],color="red", label="Pd=0.95, 2 SP")
plt.plot(df["time"], df["true count"],"-.",color="black", label="true count")
plt.xlabel("Time")
plt.ylabel("Number of targets")
plt.yticks([0,1,2])
plt.title("Number of targets in time")
plt.legend()
plt.savefig("2_sp.png",bbox_inches='tight')
plt.show()
```
